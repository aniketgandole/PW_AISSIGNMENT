```python
# Question no 1
# Discuss string slicing and provide examples>>>>>>>>>>
String slicing is a method to extract a portion of a string using
a specific range of indices. It allows you to get a substring by 
specifying a start and end index.

Syntax:string[start:end:step]
example:
text = "Hello, World!"

# Slice from index 1 to 5
slice1 = text[1:5]  # "ello"

# Slice from index 0 to 5 with step 2
slice2 = text[0:5:2]  # "Hlo"

# Slice from the start to index 5
slice3 = text[:5]  # "Hello"

# Slice from index 7 to the end
slice4 = text[7:]  # "World!"

# Reverse the string
slice5 = text[::-1]  # "!dlroW ,olleH"
```

```python
# Question no 2
#Explain the key features of lists in Python>>>>
>>>>>
Lists in Python are dynamic arrays that can hold elements 
of different data types. They are mutable, ordered, and 
allow duplicate elements.

Key Features:
Ordered: Elements have a defined order, and this order will
not change unless explicitly done.
Mutable: You can change, add, or remove elements after the
list has been created.
Heterogeneous: Lists can contain elements of different data types.
Dynamic: Lists can grow and shrink in size as needed.

example:
# Creating a list
my_list = [1, 2, 3, "four", 5.0]

# Appending an element
my_list.append(6)  # [1, 2, 3, "four", 5.0, 6]

# Removing an element
my_list.remove("four")  # [1, 2, 3, 5.0, 6]

# Accessing elements
first_element = my_list[0]  # 1

# Slicing the list
sub_list = my_list[1:4]  # [2, 3, 5.0]
```

```python
# Question no 3
#Describe how to access,modify and delete elements in a list with examples>>>>>>>>
>>>>
example:
my_list = [10, 20, 30, 40, 50]
# Accessing by index
first_element = my_list[0]  # 10
last_element = my_list[-1]  # 50
# Accessing by slicing
sub_list = my_list[1:4]  # [20, 30, 40]

1.Modifying Elements:
my_list = [10, 20, 30, 40, 50]

# Modifying by index
my_list[2] = 35  # [10, 20, 35, 40, 50]

# Modifying by slicing
my_list[1:3] = [25, 30]  # [10, 25, 30, 40, 50]

2.Deleting Elements:
my_list = [10, 20, 30, 40, 50]

# Deleting by index
del my_list[1]  # [10, 30, 40, 50]

# Deleting by slicing
del my_list[1:3]  # [10, 50]

# Removing by value
my_list.remove(10)  # [50]

# Removing by pop (returns the removed element)
last_element = my_list.pop()  # last_element = 50, my_list = []

```

```python
# Question no 4
#Compare and contrast tuples and lists with examples
>>>>>>
Tuples:

1.Immutable: Once created, elements cannot be modified.
2.Ordered: Elements have a defined order.
3.Heterogeneous: Can contain different data types.
4.Dynamic: Size is fixed once created.

Lists:

1.Mutable: Elements can be modified after creation.
2.Ordered: Elements have a defined order.
3.Heterogeneous: Can contain different data types.
4.Dynamic: Size can change as elements are added or removed.

example:
# Tuple
my_tuple = (1, 2, 3, "four")

# List
my_list = [1, 2, 3, "four"]

# Accessing elements (same syntax)
first_tuple_element = my_tuple[0]  # 1
first_list_element = my_list[0]  # 1

# Modifying elements (only possible in list)
my_list[1] = "two"  # [1, "two", 3, "four"]

# Adding elements (only possible in list)
my_list.append(5)  # [1, "two", 3, "four", 5]

# Attempt to modify tuple (raises error)
# my_tuple[1] = "two"  # TypeError: 'tuple' object does not support item assignment
```

```python
# Question no 5
# Describe the key features of sets and provide examples of their use
>>>>>>
Sets in Python are unordered collections of unique elements.

Key Features:

Unordered: Elements have no specific order.
Unique: Duplicate elements are not allowed.
Mutable: Elements can be added or removed.
Dynamic: Size can change as elements are added or removed.

Examples:
# Creating a set
my_set = {1, 2, 3, 4, 4}  # {1, 2, 3, 4}

# Adding an element
my_set.add(5)  # {1, 2, 3, 4, 5}

# Removing an element
my_set.remove(3)  # {1, 2, 4, 5}

# Checking membership
is_in_set = 2 in my_set  # True

# Set operations
another_set = {4, 5, 6, 7}
intersection = my_set & another_set  # {4, 5}
union = my_set | another_set  # {1, 2, 4, 5, 6, 7}
difference = my_set - another_set  # {1, 2}
```

```python
# Question no 6
#Discuss the use cases of tuples and sets in Python programming
>>>>>>>
Tuples:

Data Integrity: Use tuples to store related data that should not change,
such as coordinates, database records, or configurations.
Dictionary Keys: Tuples can be used as keys in dictionaries due to their
immutability.

Example:
# Coordinates
coordinates = (40.7128, 74.0060)

# Dictionary keys
my_dict = {(1, 2): "point1", (3, 4): "point2"}


Sets:

Membership Testing: Use sets for fast membership tests, like checking if
an item exists.
Mathematical Operations: Useful for mathematical operations such 
as union, intersection, and difference.
Removing Duplicates: Convert a list to a set to remove duplicates.

Example:
# Fast membership testing
my_set = {1, 2, 3, 4, 5}
exists = 3 in my_set  # True

# Removing duplicates from a list
my_list = [1, 2, 2, 3, 4, 4, 5]
unique_elements = list(set(my_list))  # [1, 2, 3, 4, 5]

```

```python
# Question no 7
#Describe How to Add, Modify, and Delete Items in a Dictionary with Examples
>>>>>>>>>>....
example:
Adding Items:
my_dict = {"name": "John", "age": 30}

# Adding a new key-value pair
my_dict["city"] = "New York"  # {"name": "John", "age": 30, "city": "New York"}


Modifying Items:
my_dict = {"name": "John", "age": 30}

# Modifying an existing key-value pair
my_dict["age"] = 31  # {"name": "John", "age": 31}


Deleting Items:
my_dict = {"name": "John", "age": 30, "city": "New York"}

# Deleting a key-value pair
del my_dict["city"]  # {"name": "John", "age": 30}

# Using pop (returns the value)
age = my_dict.pop("age")  # age = 30, my_dict = {"name": "John"}

```
